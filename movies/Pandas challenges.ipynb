```python
import pickle

with open('recent_movie_data.pkl', 'rb') as infile:
    movie_data = pickle.load(infile)
```

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
import pprint
pprint.pprint(movie_data[:10])
```

```python
mdf = pd.DataFrame(movie_data)
#mdf = mdf.dropna(0)
#len(mdf)
mdf.dtypes
```

```python
mdf_with_runtimes = mdf[-np.isnan(mdf['runtime'])]

mdf_with_runtimes['runtimes'] = mdf_with_runtimes['runtime'].astype(np.int64)

mdf_with_runtimes.dtypes
```

```python
# Challenge 1: plot domestic total gross over time
domestic_total_gross = mdf['domestic_total_gross']
dates = mdf['release_date']
date_and_incomes = sorted(zip(dates, domestic_total_gross))
date_plot = []
total_gross_plot = []
for movie in date_and_incomes:
    date_plot.append(movie[0])
    total_gross_plot.append(movie[1])

#date_plot = [date for date in dates] #otherwise pyplot won't label x-axis correctly
plt.figure(figsize=(10, 7))
plt.plot(date_plot, total_gross_plot, 'b-')
```

```python
# Challenge 1: plot domestic total gross over time
fig = plt.figure(figsize=(15, 5))
mdf = mdf.sort(columns=['release_date'])
mdf.plot(x='release_date', y='domestic_total_gross')
```

```python
# Challenge 2: plot runtime vs domestic total gross
fig = plt.figure(figsize=(15,5))
mdf = mdf.sort(columns=['domestic_total_gross'])
mdf.plot(x='domestic_total_gross', y='runtime', kind='scatter')
```

```python
# Challenge 3: group your data by Rating and find the average runtime and domestic total gross at each level of Rating
# Challenge 4: make one figure with (N=the number of MPAA ratings there are) subplots, and in each plot the release 
#              date vs the domestic total gross
grouped_by_rating = mdf.groupby('rating')

#print grouped_by_rating.describe()
#print grouped_by_rating.groups
#grouped_by_rating.plot(x='domestic_total_gross', y='runtime')
#grouped_by_rating.plot()

#plt.figure()
for letter, grp in grouped_by_rating:
    plt.subplots()
    grp.plot(x='release_date', y='domestic_total_gross').set_title(letter)

#ratings = mdf[mdf['rating'] == 'PG-13']
```

```python
# Challenge 5: what director in your dataset has the highest gross per movie?
grouped_by_director = mdf.groupby('director')
grouped_by_director.mean().sort('domestic_total_gross').tail()
#sorted(grouped_by_director.size(), key=lambda director: director[1])

```

```python
# Challenge 6: bin your dataset into years (if applicable) and make a bar graph with error bars of gross each year

grouped_by_year = mdf.groupby(mdf['release_date'].map(lambda x: x.year))
grouped_by_year.size()
```

```python
# Challege 6
stddevs = grouped_by_year.std()['domestic_total_gross']
grouped_by_year.mean()['domestic_total_gross'].plot(kind='bar', yerr=stddevs)
```
