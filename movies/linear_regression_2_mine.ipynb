```python
%matplotlib inline
import pandas as pd
import numpy as np
import pickle
import random
import statsmodels
import matplotlib.pyplot as plt
from datetime import datetime

import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
# My data
#with open('recent_movie_data.pkl', 'rb') as infile:
#    movie_data = pickle.load(infile)
#mdf = pd.DataFrame(movie_data)

# Irmak's data
mdf = pd.read_csv('2013_movies.csv', parse_dates={'Date':['ReleaseDate']})

mdf.head()
```



Challenge 1

Instead of predicting total gross directly, build a linear model that predicts the 
log (logarithm) of the total gross. Plot the model. Plot the predictions versus 
actual values. Does this look like a better model?

```python
movies = mdf.dropna()
movies['Ones'] = 1.0
print len(movies)
movies.head()
```

```python
# Irmak's data
movies['domestic_total_gross_log'] = np.log(movies['DomesticTotalGross'])

# My data
#movies['domestic_total_gross_log'] = np.log(movies['domestic_total_gross'])
movies.head()
```

```python
X = movies[['Budget', 'Ones']]
Y = movies['domestic_total_gross_log']
linmodel_log_gross = sm.OLS(Y, X).fit()
linmodel_log_gross.summary()
```

```python
plt.scatter(movies.Budget, movies.domestic_total_gross_log)
plt.xlabel('Budget')
plt.ylabel('log(Domestic Total Gross')
plt.plot(movies.Budget, linmodel_log_gross.predict(X), 'r')

prstd, conf_lower, conf_upper = wls_prediction_std(linmodel_log_gross)
plt.plot(movies.Budget, conf_lower, 'y--')
plt.plot(movies.Budget, conf_upper, 'y--')
```

```python
# plot the predictions 
plt.plot(movies.domestic_total_gross_log, linmodel_log_gross.predict(), 'go')
# plot the 1:1 actual gross line
plt.plot(movies.domestic_total_gross_log, movies.domestic_total_gross_log, 'b--')
# axis labels
plt.xlabel('Actual log(Gross)')
plt.ylabel('Predicted log(Gross)')
```

Build a model that uses average director gross as a feature (one of the predictor variables). 
Fit and evaluate. Would you use this model to predict budgets of upcoming movies?

```python
movies = movies.join(movies.groupby('Director')['DomesticTotalGross'].mean(), on='Director', rsuffix='_director_mean')
```

```python
# Randomly select 75% of movies and split into two dataframes
rows = random.sample(movies.index, int(0.75*len(movies)))

movies_75 = movies.ix[rows]
movies_25 = movies.drop(rows)
print '75% of movies:', len(movies_75)
print '25% of movies:', len(movies_25)
#print len(movies.index)
print 'Total number of movies:', len(movies)


X = movies_75[['Budget', 'DomesticTotalGross_director_mean', 'Ones']]
Y = movies_75['DomesticTotalGross']
linmodel_log_gross_75_dirmean = sm.OLS(Y, X).fit()
linmodel_log_gross_75_dirmean.summary()
```

```python
X_25 = movies_25[['Budget', 'DomesticTotalGross_director_mean', 'Ones']]
# plot the predictions 
plt.plot(movies_25.DomesticTotalGross, linmodel_log_gross_75_dirmean.predict(X_25), 'go')
# plot the 1:1 actual gross line
plt.plot(movies_25.DomesticTotalGross, movies_25.DomesticTotalGross, 'b--')
# axis labels
plt.xlabel('Actual log(Gross)')
plt.ylabel('Predicted log(Gross)')
plt.title('Linear Model using Director Gross Mean')
```
