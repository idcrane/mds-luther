```python
# Linear Model and Analysis functions

def run_country_analysis(country_data, traindata, testdata, Xcolumn_list, Ycolumn):
    X_train = traindata[Xcolumn_list]
    Y_train = traindata[Ycolumn]
    X_test = testdata[Xcolumn_list]
    Y_test = testdata[Ycolumn]

    # Do one linear model
    linmodel = sm.OLS(Y_train, X_train).fit()
#     print linmodel.summary() 
    Y_predict = linmodel.predict(X_test)

    # Do cross_validation and get errors
    linmodel, mse_test = do_linmodel(traindata, testdata, Xcolumn_list, Ycolumn)
    mse_error_list, rsquared_list = errors_vs_num_elements(traindata, testdata, Xcolumn_list, Ycolumn)
    mean_mse_errors = cross_validation(country_data, Xcolumn_list, Ycolumn)                             
    mean_mse_std = np.sqrt(np.array(mean_mse_errors))
    print 'Mean MSE Std =', mean_mse_std
    
    data_dict = {'X_train': X_train, 'y_train': Y_train, 'X_test': X_test, 'y_test':Y_test,
                 'y_predict': Y_predict, 'mean_mse_std': mean_mse_std}

    return linmodel, data_dict

def do_linmodel(traindata, testdata, Xcolumn_list, Ycolumn, print_summary=True):
    """ Do linmodel testing, find mean squared errors 
    """
    X_train = traindata[Xcolumn_list]
    Y_train = traindata[Ycolumn]
    X_test = testdata[Xcolumn_list]
    Y_test = testdata[Ycolumn]
        
    linmodel = sm.OLS(Y_train, X_train).fit()
    if print_summary:
        print linmodel.summary() 
    Y_predict = linmodel.predict(X_test)
    mse_test = mean_squared_error(Y_test, Y_predict) 
    return linmodel, mse_test

def errors_vs_num_elements(traindata, testdata, Xcolumn_list, Ycolumn):
    """Loop through linmodel testing, adding element one at a time
    """
    mse_error_list = []
    rsquared_list = []
    Xcolumn_list_list = []
    for i in range(2, len(Xcolumn_list) + 1):
        Xcolumn_list_list.append(Xcolumn_list[0:i])
        linmodel, mse_test = do_linmodel(traindata, testdata, Xcolumn_list[0:i], Ycolumn, print_summary=False)
        mse_error_list.append(mse_test)
        rsquared_list.append(linmodel.rsquared)
    return mse_error_list, rsquared_list #, Xcolumn_list_list

def cross_validation(dataframe, Xcolumn_list, Ycolumn, numtrials=50):
    """ Do cross validation
    """
    master_mse_error_list = []
    master_rsquared_list = []
    for i in range(numtrials):
        movies_75, movies_25 = random_sample_DataFrame(dataframe)
        mse_error_list, rsquared_list = errors_vs_num_elements(movies_75, movies_25, Xcolumn_list, Ycolumn)
        master_mse_error_list.append(mse_error_list)
        master_rsquared_list.append(rsquared_list)
    mean_mse_errors = np.mean(np.array(master_mse_error_list), axis=0)
    return mean_mse_errors
```

```python
# Create random sample defintion
def random_sample_DataFrame(dataframe, proportion=0.75):
    rows = random.sample(dataframe.index, int(proportion*len(dataframe)))
    movies_75 = dataframe.ix[rows]
    movies_25 = dataframe.drop(rows)
    return movies_75, movies_25

# Create float time deltas when np.nan's are in same data
def change_delta_times(datetimeobject):
    if isinstance(datetimeobject, float):
        return np.nan
    if isinstance(datetimeobject, datetime.timedelta):
        return float(datetimeobject.days)
 
def do_linmodel_notraining(data, Xcolumn_list, Ycolumn, print_summary=True): 
    Y = data[Ycolumn]
    X = data[Xcolumn_list]
        
    linmodel = sm.OLS(Y, X).fit()
    if print_summary:
        print linmodel.summary() 
    Y_predict = linmodel.predict(X)
    mse_test = mean_squared_error(Y, Y_predict) 
    return linmodel, mse_test    

# Randomly choose some movies to plot
def plot_point_select(data, numpoint=4):
    x_100 = []
    x_200 = []
    x_300 = []
    x_400 = []
    x_500 = []
    for s in data:
        x = float(s[0])
#         print x
        if x > float(0) and x < float(100):
            x_100.append(s)
        elif x > float(100) and x <= float(200):
            x_200.append(s)
        elif x > float(200) and x <= float(300):
            x_300.append(s)
        elif x > float(300) and x <= float(400):
            x_400.append(s)
        else:
            x_500.append(s)        
    if numpoint > len(x_100):
        numpoint = len(x_100)
    x_100_out = random.sample(x_100, numpoint)
    if numpoint > len(x_200):
        numpoint = len(x_200)
    x_200_out = random.sample(x_200, numpoint)
    if numpoint > len(x_300):
        numpoint = len(x_300)
    x_300_out = random.sample(x_300, numpoint)
    if numpoint > len(x_400):
        numpoint = len(x_400)
    x_400_out = random.sample(x_400, numpoint)
    if numpoint > len(x_500):
        numpoint = len(x_500)
    x_500_out = random.sample(x_500, numpoint)    
    x_plot = x_100_out + x_200_out + x_300_out + x_400_out +x_500_out
    return x_plot

def mannwhitney_test(country_data, country_gross_field_name, sample_size=30):
    """ Do Mann-Whitney Test to see if presence of in-country actors has 
        statistical effect on in-country gross.
    """
    country_data_noactor = country_data[country_data['asian_actors'] == 0]
    country_data_actor = country_data[country_data['asian_actors'] > 0]
    print '======================================================================'
    print 'Mann-Whitney Test of In-Country Actor Number Influence'
    print 'Movies with in-country actors:', len(country_data_actor)
    print 'Movies with NO in-country actors', len(country_data_noactor)

    if sample_size > len(country_data_actor):
        sample_size = len(country_data_actor) + 10
    
    rows = random.sample(country_data_noactor.index, sample_size)
    country_data_noactor_sample = country_data_noactor.ix[rows]
    
    noactor_list = list(country_data_noactor_sample[country_gross_field_name])
    actor_list = list(country_data_actor[country_gross_field_name])
    
    u, prob = mannwhitneyu(actor_list, noactor_list)
    print 'Mean no actor ($ Million) =', np.mean(np.array(noactor_list))/1e6
    print 'Mean actor ($ Million) =', np.mean(np.array(actor_list))/1e6
    print 'u =', u
    print 'prob =', prob
    print '\n======================================================================'
    print '\n\n'
```

```python
# Make plots uniform
def plot_error_vs_features(y_values, x_tick_labels, title, figwidth=12, figheight=8, force_ylim=False,
                           force_xlim=False, ylim=[40, 50], xlim=[40, 50]):
    assert len(y_values) == len(x_tick_labels)
    fig, ax = plt.subplots(1)
    x_dummy = range(len(mean_mse_std))
    fig.set_figwidth(figwidth)
    fig.set_figheight(figheight)
    ppl.bar(ax, x_dummy, y_values, align='center')
    if force_ylim:
        ax.set_ylim(ylim)
    if force_xlim:
        ax.set_xlim(xlim)
    ax.set_ylabel('Prediction Error ($ Million)', fontweight='bold', fontsize=18)
    ax.set_xlabel('Elements Used in Model', fontweight='bold', fontsize=18)
    plt.tick_params(labelsize=16)
    ax.set_xticks(x_dummy)
    ax.set_xticklabels(x_tick_labels, fontweight='bold', fontsize=16)
    ax.set_title(title, fontweight='bold', fontsize=20)
    
def plot_predicted_and_actual(x_values, y_test, y_predict, mean_mse_std, ylabel, title, figwidth=12, figheight=8, force_ylim=True,
                           force_xlim=False, ylim=[-10, 170], xlim=[40, 50]):
    fig, ax = plt.subplots(1)    
    fig.set_figwidth(figwidth)
    fig.set_figheight(figheight)
    ppl.plot(x_values,y_test, 'ro', label='Actual Gross')
    plt.errorbar(x_values, y_predict, yerr=2*mean_mse_std, fmt='bo', label='Predicted Gross')
    if force_ylim:
        ax.set_ylim(ylim)
    if force_xlim:
        ax.set_xlim(xlim)
    ax.set_ylabel(ylabel, fontweight='bold', fontsize=18)
    ax.set_xlabel('US Total Gross ($ Million)', fontweight='bold', fontsize=18)
    plt.tick_params(labelsize=16)
    ax.set_title(title, fontweight='bold', fontsize=20)
    plt.legend()
    plt.legend(loc=2, fontsize=16)
```
